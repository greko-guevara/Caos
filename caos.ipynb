```python
#codigo de entrada
print ("***********************************************************")
print ("Simulador de ecuaci√≥n de caos para diferentes valores de r")
print ("ver https://youtu.be/EOvLhZPevm0?si=UsFhCv1dxiQsu5Ci")
n=  int (50)
x= 0.25
xx=x 

import pandas as pd
import numpy as np 
df=pd.DataFrame()
df_con_rr=pd.DataFrame()
df_con_x=pd.DataFrame()


r=[0.5,1,1.5,2,2.5,3,3.25,3.5,3.75,3.9]
a=range(1,n)
grafico=-1
import matplotlib.pyplot as plt
import matplotlib
fig, ax = plt.subplots(2, 5, figsize=(15, 10))
plt.grid(True)
for k in r:
    grafico=grafico+1
    for j in a:
        df.at[j,'n']=j
        df.at[j,'x']=x
        x=k*x*(1-x)
        df.at[j,'x+1']=x
    if grafico<5:
        ax[0,grafico].plot(df["n"], df["x+1"])
        ax[0,grafico].set_title(k)
    else:
        grafico1=grafico-5
        ax[1,grafico1].plot(df["n"], df["x+1"])
        ax[1,grafico1].set_title(k)
    x=xx

plt.show()
```

```python

rr=0
x=xx
while rr<4:
    rr=rr+0.0001
    x=rr*x*(1-x)
    df_con_rr.append(rr)
    df_con_x.append(x)

fig, ax = plt.subplots( figsize=(10, 10))
ax.plot(df_con_rr, df_con_x)
plt.show()
```
